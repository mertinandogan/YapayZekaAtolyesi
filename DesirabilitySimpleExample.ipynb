```python
def min_based(y, h_min, h_max, r):
    if h_min >= y:
        return 1
    if y <= h_max and h_min <= y:
        return ((y-h_max) / (h_min - h_max) )**r
    if y >= h_max:
        return 0
    
def max_based(y, h_min, h_max, r):
    if h_min >= y:
        return 0
    if y <= h_max and h_min <= y:
        return ((y-h_min) / (h_max - h_min) )**r
    if y >= h_max:
        return 1
    
def nom_based(y, h_min, h_med, h_max, s, t):
    if h_min >= y:
        return 0
    if y <= h_med and h_min <= y:
        return ((y-h_min) / (h_med - h_min) )**t
    if y >= h_med and h_max >= y:
        return ((y-h_max) / (h_med - h_max) )**s
    if y >= h_max:
        return 0
    
def func1( w, l , t, eps):
    return ((w + l) + (t - eps) ) / 10**2

def func2( w, l , t, eps):
    return ((w - l) + (t + eps) ) / 10**2

def func3( w, l , t, eps):
    return ((w - l) + (t - eps) ) / 10**2

def dw(y, h_min, h_max, r):
    return min_based( y, h_min, h_max, r)

def dl(y, h_min, h_max, r):
    return min_based( y, h_min, h_max, r)

def dt(y, h_min, h_max, r):
    return min_based( y, h_min, h_max, r)

def deps(y, h_min, h_med, h_max, s, t):
    return nom_based( y, h_min, h_med, h_max, s, t)

def dwl(y, h_min, h_med, h_max, s, t):
    return nom_based( y, h_min, h_med, h_max, s, t)

def dg(y, h_min, h_max, r):
    return max_based( y, h_min, h_max, r)

def dr(y, h_min, h_max, r):
    return min_based( y, h_min, h_max, r)

def df(y, h_min, h_max, r):
    return min_based( y, h_min, h_max, r)

def d_overall(w, w_min, w_max, r_w, 
              l, l_min, l_max, r_l,
              t, t_min, t_max, r_t,
              eps, eps_min, eps_med, eps_max, r_eps_1, r_eps_2,
              w_over_l_min, w_over_l_med, w_over_l_max, r_w_over_l_1, r_w_over_l_2,   
              func1_min, func1_max, r_func1,
              func2_min, func2_max, r_func2,
              func3_min, func3_max, r_func3):
    
    res1 = dw(w, w_min, w_max, r_w)
    res2 = dl(l, l_min, l_max, r_l)
    res3 = dt(t, t_min, t_max, r_t)
    res4 = deps(eps, eps_min, eps_med, eps_max, r_eps_1, r_eps_2)
    res5 = dwl(w/l, w_over_l_min, w_over_l_med, w_over_l_max, r_w_over_l_1, r_w_over_l_2)
    res6 = dg(func1(w, l , t, eps), func1_min, func1_max, r_func1)
    res7 = dr(func2(w, l , t, eps), func2_min, func2_max, r_func2)
    res8 = df(func3(w, l , t, eps), func3_min, func3_max, r_func3)

    overall_desirability = (res1*res2*res3*res4*res5*res6*res7*res8)**(1/8)
    return overall_desirability

# w = 9
# l = 8
# t = 0.8
# eps = 5.4

# res1 = dw( w, 0, 10, 1)
# res2 = dl( l, 0, 10, 1)
# res3 = dt( t, 0, 1, 10)

# res4 = deps(eps, 1, 5.6, 8, 1, 1.1)
# res5 = dwl(w/l, 1/1.5, 1, 1.5, 0.1, 0.1)
# res6 = dg( func1( w, l , t, eps), 0, 1, 1)
# res7 = dr( func2( w, l , t, eps), 0, 1, 1)
# res8 = df( func3( w, l , t, eps), 0, 1, 1)
# 
# overall_res = (res1*res2*res3*res4*res5*res6*res7*res8)**(1/8)
# print(overall_res)
```

```python
import numpy as np

def izgaraArama( l_alt, l_ust, w_alt, w_ust, t_alt, t_ust, eps_alt, eps_ust, 
                 l_adim, w_adim, t_adim, eps_adim):
    arr_l = np.arange(l_alt, l_ust, l_adim)
    arr_w = np.arange(w_alt, w_ust, w_adim)
    arr_t = np.arange(t_alt, t_ust, t_adim)
    arr_eps = np.arange(eps_alt, eps_ust, eps_adim)
    
    w_min = w_alt 
    w_max = w_ust 
    r_w = 1
    l_min = l_alt
    l_max = l_ust 
    r_l = 1
    t_min = t_alt 
    t_max = t_ust 
    r_t = 10
    eps_min = eps_alt 
    eps_med = 5.6 
    eps_max = eps_ust 
    r_eps_1 = 1 
    r_eps_2 = 1.1
    w_over_l_min = 1/1.5
    w_over_l_med = 1 
    w_over_l_max = 1.5 
    r_w_over_l_1 = 0.1
    r_w_over_l_2 = 0.1   
    func1_min = 0 
    func1_max = 1 
    r_func1 = 1
    func2_min = 0 
    func2_max= 1 
    r_func2 = 1
    func3_min = 0 
    func3_max = 1 
    r_func3 = 1
    
    dl_values = []
    dw_values = []
    dt_values = []
    deps_values = []
    dlw_values = []
    dfunc1_values = []
    dfunc2_values = []
    dfunc3_values = []
    d_overall_values = []
        
    for i in arr_l:
        for j in arr_w:
            for k in arr_t:
                for z in arr_eps:
                    trial_result = d_overall(j, w_min, w_max, r_w, 
                                             i, l_min, l_max, r_l,
                                             k, t_min, t_max, r_t,
                                             z, eps_min, eps_med, eps_max, r_eps_1, r_eps_2,
                                             w_over_l_min, w_over_l_med, w_over_l_max, r_w_over_l_1, r_w_over_l_2,
                                             func1_min, func1_max, r_func1,
                                             func2_min, func2_max, r_func2,
                                             func3_min, func3_max, r_func3)
                    d_overall_values.append(trial_result)
            
    return d_overall_values 

d_overall_values_array = izgaraArama(0.1, 10, 0, 10, 0, 1, 1, 8, 1, 1, 0.1, 0.5)
# plt.scatter(f1,f2)
```
