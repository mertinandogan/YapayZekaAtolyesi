```python
# pulp function
def Do_Pulp_Classification (tumorData, threshold):
    result_array = []
    malign = 1
    benign = 0
    
    for i in range(len(tumorData["Hasta ID"])):
        a = tumorData["Tümör Yarıçapı (mm)"].iloc[i]
        b = tumorData["Bağ Dokusu (Yok = 0; Var = 1)"].iloc[i]
        if a >= threshold and b == 1:
            result_array.append(malign)
        else:
            result_array.append(benign)
    
    return result_array
```

```python
result = Do_Pulp_Classification(tumorData, 40)
sb.histplot(result, kde= True)
```

```python
def Do_Pulp_Classification_With_Lambda (tumorData, threshold):
    result_array = []
    tumorData["Out_put"]= tumorData["Tümör Yarıçapı (mm)"].apply(lambda x : 1 if x>= threshold else 0)
    return tumorData["Out_put"]
```

```python
result = Do_Pulp_Classification_With_Lambda(tumorData, 40)
sb.histplot(result, kde= True)
```

```python
def Do_Pulp_Classification_With_Less_Code(tumorData, threshold):
    return np.array(tumorData["Tümör Yarıçapı (mm)"]>=threshold,dtype=np.float)
```

```python
result = Do_Pulp_Classification_With_Less_Code(tumorData,40)
sb.histplot(result,kde=True)
plt.show()
```

```python
def Evaluate_Performance(tumorData, result_array):
    number_of_true_positives = 0
    number_of_true_negatives = 0
    number_of_false_positives = 0
    number_of_false_negatives = 0
    
    for i in range(len(tumorData["Hasta ID"])):
        truth = tumorData["Etiket (İyi Huylu = 0; Kötü Huylu = 1)"].iloc[i]
        response = result_array[i]
        if truth == 1 and response == 1:
            number_of_true_positives +=1
        elif truth == 1 and response == 0:
            number_of_false_negatives +=1
        elif truth == 0 and response == 0:
            number_of_true_negatives +=1
        else:
            number_of_false_positives +=1
        
    tp = number_of_true_positives
    tn = number_of_true_negatives
    fp = number_of_false_positives
    fn = number_of_false_negatives
    
    sensitivity = tp/(tp+fn)
    
    specificity = tn/(tn+fp)
    
    accuracy = (tp+tn)/(tp+tn+fp+fn)
    
    precision = tp/(tp+fp)
    
    f1_score = (2*precision*sensitivity)/(precision+sensitivity)
    
    false_positive_rate = 1-specificity
    
    print('Sensitivity: {}, Specificity: {}, Accuracy {}, precision {}, f1_score {}, false_positive_rate {}'.format(sensitivity,specificity,accuracy,precision,f1_score,false_positive_rate))
    
    return specificity, sensitivity, accuracy, precision, f1_score, false_positive_rate
```

```python
Evaluate_Performance(tumorData, result)
```

```python
def Do_ROC_Analysis(tumorData):
    threshold_upper_bound = 100
    array1 = []
    array2 = []
    
    for i in range(0,threshold_upper_bound,10):
        
        result = Do_Pulp_Classification_With_Less_Code(tumorData, i)
        specificity, sensitivity, accuracy, precision, f1_score, false_positive_rate = Evaluate_Performance(tumorData, result)
        
        array1.append(1-specificity)
        array2.append(sensitivity)
        plt.plot((1-specificity), sensitivity, "o")
    plt.plot(np.array(array1), np.array(array2))
        
        
```

```python
Do_ROC_Analysis(tumorData)
plt.show()
```

```python
def Do_ROC_Analysis_With_2_Param(tumorData):
    threshold_upper_bound = 100
    array1 = []
    array2 = []
    
    for i in range(0,threshold_upper_bound,10):
        
        result = Do_Pulp_Classification(tumorData, i)
        specificity, sensitivity, accuracy, precision, f1_score, false_positive_rate = Evaluate_Performance(tumorData, result)
        
        array1.append(1-specificity)
        array2.append(sensitivity)
        plt.plot((1-specificity), sensitivity, "o")
    plt.plot(np.array(array1), np.array(array2))
    
    return plt
```
