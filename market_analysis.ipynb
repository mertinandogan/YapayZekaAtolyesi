```
rawDS = spark.read.option("delimiter", ";").option("header",True).option("inferSchema", "true").csv("/FileStore/tables/market2.csv")
```

```
rawDS.show(truncate=False)
```

```
rawDS.printSchema()
```

```
selectedDS = rawDS.select("ITEMNAME","AMOUNT","PRICE","CITY","CLIENTNAME","ITEMCODE")
```

```
selectedDS.show(truncate = False)
```

```
city_countDS = selectedDS.groupBy("CITY").count()
```

```
selectedDS.groupBy("CLIENTNAME").count().show()
```

```
renamedDS = city_countDS.withColumnRenamed("count", "CITY_COUNT")
```

```
renamedDS.filter("CITY_COUNT >1000").show()
```

```
renamedDS.filter(renamedDS["CITY_COUNT"] >1000).show()
```

```
renamedDS.filter("CITY_COUNT >1000 and CITY_COUNT < 1500").show()
```

```
rawDS.groupBy("CLIENTCODE").sum("LINENETTOTAL").orderBy("sum(LINENETTOTAL)", ascending = False).show()
```

```
from pyspark.sql.functions import *
```

```
clientCodeTotalDS = rawDS.groupBy("CLIENTCODE").sum("LINENETTOTAL")
```

```
finalDS = clientCodeTotalDS.withColumn("TOTAL_SPEND", round("sum(LINENETTOTAL)", 2)).drop("sum(LINENETTOTAL)")
```

```
finalDS.show(truncate=False)
```

```
customerTop5 = finalDS.orderBy("TOTAL_SPEND", ascending = False)
```

```
customerTop5.show(5)
```

```
customerTop5.na.drop().show(5)
```

```
customerTop5.na.fill({"CLIENTCODE":-1}).show(5)
```
